```python
#compute the tf idf and tf-idf values for the starting word in the text

#computeTF
def TF(sentence):
  word = sentence.split(" ")
  value = len(set(word))
  return words.count(word[0])/value
  pass
```

```python
#compute IDF

import math 

def IDF(sentence):
  words = sentence.lower().split(" ")
  res = words.count(words[0])
  return math.log(len(words)/res, 10)
  pass
```

```python

lemmatized = required_df1["final_data"].values.tolist()
pos=[]
count=0
for i in lemmatized:
  text = word_tokenize(i)
  count=count+1
  pos.append(nltk.pos_tag(text))    #Parts of Speech for each sentence
```

```python
from collections import Counter

def get_pos(value,position):
  pos_counts=[]
  for j in position:
    counts = Counter(tag for word,tag in j)  
    if(counts.get(value)!=None):
      pos_counts.append(counts.get(value))
    else:
      pos_counts.append(0)
  return pos_counts
```

```python
required_df1["#nouns"] = pd.DataFrame(get_pos("NN",pos))
required_df1["#adverb"] = pd.DataFrame(get_pos("RB",pos))
required_df1["#cardinal digits"] = pd.DataFrame(get_pos("CD", pos))
required_df1["#Verbs"] = pd.DataFrame(get_pos("VB",pos))
required_df1["#Determiners"] = pd.DataFrame(get_pos("DT",pos)) 
required_df1["#adjectives"] = pd.DataFrame(get_pos("JJ",pos)) 
required_df1["#interjenctions"] = pd.DataFrame(get_pos("UH",pos))
required_df1["#Prepositions"] = pd.DataFrame(get_pos("IN",pos))
```

```python
#find the pronouns
one = pd.DataFrame(get_pos("PRP",pos))
two = pd.DataFrame(get_pos("PRP$",pos))
three = pd.DataFrame(get_pos("WP",pos))
four = pd.DataFrame(get_pos("WP$",pos))

required_df1["number of pronouns"] = one[0]+two[0]+three[0]+four[0]
```

```python
one = required_df1["final_data"].apply(lambda x : len(x.split(" ")))
required_df1["Density of pronouns"] = required_df1["number of pronouns"]/one[0]
required_df1["Density of nouns"] = required_df1["Number of nouns"]/one[0]
required_df1["#conjunctions"] = pd.DataFrame(get_pos("CC",pos))
required_df1["Density of prepositions"] = required_df1["Number of Prepositions"]/one[0]
required_df1["Density of conjunctions"] = required_df1["Number of conjunctions"]/one[0]
```
