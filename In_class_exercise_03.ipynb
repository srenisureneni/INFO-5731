```python
# You code here (Please add comments in the code):

#Import pandas and NLTK library 
import pandas as pd
import nltk 
from nltk.tokenize import word_tokenize  

#Read the required dataset using Pandas
df = pd.read_csv("Crime_Data_2010_2017 (1).csv")
print(df.head())
```

```python
#Assign the Crime Code Description column as list 
word_tokens = list(df["Crime Code Description"])

#Use FreqDist function in NLTK library
freq = nltk.FreqDist(word_tokens)

#Convert the variable into dictionary
dict_freq = dict(freq)
print(dict_freq)

#Plot the top 10 Crime Codes description
freq.plot(10, cumulative=False)
```
